```python
def soma(a,b):
  return a+b 
def subtr(a,b):
  return a-b
def mult(a,b):
  return a*b 
def div(a,b):
  return a/b
oper={
    '+':soma,
    '-':subtr,
    '*':mult,
    '/':div}
```

```python
def calcul():
  print(logo)
  continua=True
  n1=float(input("First number: "))
  while continua:
    op=input("Pick an operation: ")
    n2=float(input("Second number: "))
    resultado=oper[op](n1,n2)
    print(f"{n1} {op} {n2} = {resultado}")
    usr_decision=input(f"Type 'y' to continue calculating with {resultado} or type 'n' to finish...\n")
    if usr_decision=='n':
      continua=False
      #calcul()
    else:
      n1=resultado
```
